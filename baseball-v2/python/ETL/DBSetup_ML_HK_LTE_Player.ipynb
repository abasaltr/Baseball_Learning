```python
# Read Players Runs Score Predictions csv into a DataFrame
runsScoreDF = pd.read_csv(pathRunsScore)
runsScoreDF = runsScoreDF.rename(columns={'Unnamed: 0': 'ID'})
runsScoreDF = runsScoreDF.rename(columns={'model_type': 'modelType'})
runsScoreDF['ID'] = runsScoreDF.index
print(runsScoreDF.shape)
runsScoreDF.head()
```

```python
# Read csv into a DataFrame
homeRunsDF = pd.read_csv(pathHomeRuns)
homeRunsDF = homeRunsDF.rename(columns={'Unnamed: 0': 'ID'})
homeRunsDF = homeRunsDF.rename(columns={'model_type': 'modelType'})
homeRunsDF['ID'] = homeRunsDF.index
print(homeRunsDF.shape)
homeRunsDF.head()
```

```python
# Read csv into a DataFrame
baseHitsDF = pd.read_csv(pathBaseHits)
baseHitsDF = baseHitsDF.rename(columns={'Unnamed: 0': 'ID'})
baseHitsDF = baseHitsDF.rename(columns={'model_type': 'modelType'})
baseHitsDF['ID'] = baseHitsDF.index
print(baseHitsDF.shape)
baseHitsDF.head()
```

```python
# Read csv into a DataFrame
walksDF = pd.read_csv(pathWalks)
walksDF = walksDF.rename(columns={'Unnamed: 0': 'ID'})
walksDF = walksDF.rename(columns={'model_type': 'modelType'})
walksDF['ID'] = walksDF.index 
print(walksDF.shape)
walksDF.head()
```

```python
playerDF = runsScoreDF.merge(homeRunsDF[['fpID','yearID','modelType','actual','model']], on='fpID', how='outer')
print(playerDF.shape)
playerDF.head()
```

```python
playerDF3 = baseHitsDF.merge(walksDF[['fpID','yearID','modelType','actual','model']], on='fpID', how='outer')
print(playerDF3.shape)
playerDF3.head()
```

```python
lsi = []
for i in range(len(playerDF)):
    if ((playerDF.iloc[i,1] == playerDF.iloc[i,6]) and (playerDF.iloc[i,5] == playerDF.iloc[i,7])) :
        lsi.append(i)
print(len(lsi))
```

```python
lsi2 = []
for i in range(len(playerDF3)):
    if ((playerDF3.iloc[i,1] == playerDF3.iloc[i,6]) and (playerDF3.iloc[i,5] == playerDF3.iloc[i,7])) :
        lsi2.append(i)
print(len(lsi2))
```

```python
playerDF2 = playerDF.iloc[lsi,:]
playerDF2 = playerDF2[playerDF2.columns.drop('yearID_y')]
playerDF2 = playerDF2[playerDF2.columns.drop('modelType_y')]
playerDF2 = playerDF2.rename(columns={'yearID_x':'yearID'})
playerDF2 = playerDF2.rename(columns={'modelType_x':'modelType'})
playerDF2 = playerDF2.rename(columns={'actual_x':'actual_r'})
playerDF2 = playerDF2.rename(columns={'model_x':'model_r'})
playerDF2 = playerDF2.rename(columns={'actual_y':'actual_hr'})
playerDF2 = playerDF2.rename(columns={'model_y':'model_hr'})
playerDF2 = playerDF2.reset_index(drop=True)
playerDF2['ID'] = playerDF2.index
playerDF2
```

```python
playerDF4 = playerDF3.iloc[lsi2,:]
playerDF4 = playerDF4[playerDF4.columns.drop('yearID_y')]
playerDF4 = playerDF4[playerDF4.columns.drop('modelType_y')]
playerDF4 = playerDF4.rename(columns={'yearID_x':'yearID'})
playerDF4 = playerDF4.rename(columns={'modelType_x':'modelType'})
playerDF4 = playerDF4.rename(columns={'actual_x':'actual_h'})
playerDF4 = playerDF4.rename(columns={'model_x':'model_h'})
playerDF4 = playerDF4.rename(columns={'actual_y':'actual_bb'})
playerDF4 = playerDF4.rename(columns={'model_y':'model_bb'})
playerDF4 = playerDF4.reset_index(drop=True)
playerDF4['ID'] = playerDF4.index
playerDF4
```

```python
playerDF5 = pd.merge(playerDF2,playerDF4, on="fpID")
print(playerDF5.shape)
playerDF5.head()
```

```python
lsi3 = []
for i in range(len(playerDF5)):
    if ((playerDF5.iloc[i,1] == playerDF5.iloc[i,9]) and (playerDF5.iloc[i,5] == playerDF5.iloc[i,12])) :
        lsi3.append(i)
print(len(lsi3))
```

```python
playerDF6 = playerDF5.iloc[lsi3,:]
playerDF6 = playerDF6[playerDF6.columns.drop('yearID_y')]
playerDF6 = playerDF6[playerDF6.columns.drop('modelType_y')]
playerDF6 = playerDF6[playerDF6.columns.drop('ID_y')]
playerDF6 = playerDF6.rename(columns={'yearID_x':'yearID'})
playerDF6 = playerDF6.rename(columns={'modelType_x':'modelType'})
playerDF6 = playerDF6.rename(columns={'ID_x':'ID'})
playerDF6 = playerDF6.reset_index(drop=True)
playerDF6['ID'] = playerDF6.index
playerDF6
```

```python
playerPredictDF = playerDF6.copy()
playerPredictDF.head()
```
